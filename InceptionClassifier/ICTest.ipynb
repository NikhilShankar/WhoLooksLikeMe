#### **Creating another model with Direct classification approach instead of embedding based**

#### **Creating Dataset**

```python
from TrainTestSplitter import TrainTestSplitter
from DataPreparation import DataPreparation
import os

original_data_dir = "../dataset_images"
dataset_dir = "../datasets/IC-100-Basic"  # Path to your dataset folder
sample_class_number = 100

data_preparation = DataPreparation(original_data_dir, dataset_dir, sample_class_number)
selected_classes = data_preparation.prepare_data()

# Example Usage:
splitter = TrainTestSplitter(dataset_dir)
splitter.create_train_test_split()
```

```python
from InceptionClassifier import InceptionClassifier

model_builder = InceptionClassifier()
model = model_builder.build_model(num_classes=sample_class_number)  # Replace `5` with the actual number of classes
model = model_builder.compile_model(model)
```

#### **Training**

```python
from TrainAndTestIC import TrainAndTestIC

trainer = TrainAndTestIC(dataset_dir, model)
trainer.train_model(epochs=30)
trainer.evaluate_model()
```

#### **Prediction**

```python
from PredictorIC import PredictorIC

predictor = PredictorIC(model)
top_classes, scores = predictor.predict('../test_image/Adam.jpg')
print(f"Top classes: {top_classes}")
print(f"Scores: {scores}")
```

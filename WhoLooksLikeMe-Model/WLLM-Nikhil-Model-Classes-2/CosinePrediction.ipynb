#### **Predictor based on Cosine Similarity**

- For each model we have two outputs.
- One from the embedding layer
- One from the final classification layer.
- The embedding layer is used to generate embedding for a personality used during training.
- These embeddings are numpy vector or arrays.
- We can increase the number of test samples we store in the embedding to increase accuracy of this method.
- Now we have embeddings for each personality used for each model.

##### **Prediction**
- Now while predicting we generate the embedding of the image we need to predict from each of these models.
- Then we calculate cosine similarity against each personality embeddings we have stored from training.
- Combine all of those scores and we have a sorted list of personalities based on this cosine similarity.


##### **Improving prediction**
- Now instead of creating a single embedding for the image to predict we will calculate scores performing augmenting on the image to predict
- We have observed that as we increase the number of these augmentation rounds our accuracy is improved.
- The tradeoff here is between time and accuracy. So as we increase augmentation rounds time taken for prediction also goes up.

#### **Creating another model with Direct classification approach instead of embedding based**

#### **Creating Dataset**

```python
from VGGClassifier import VGGClassifier

model_builder = VGGClassifier()
dataset_dir = "../datasets/VGG-10-Basic"
model = model_builder.build_model(num_classes=10)  # Replace `5` with the actual number of classes
model = model_builder.compile_model(model)
```

#### **Training**

```python
from TrainAndTestVGG import TrainAndTestVGG

dataset_dir = "../datasets/VGG-10-Basic"
trainer = TrainAndTestVGG(dataset_dir, model)
trainer.train_model(epochs=30)
trainer.evaluate_model()
```

#### **Prediction**

```python
from PredictorVGG import PredictorVGG

predictor = PredictorVGG(model, f'{dataset_dir}/train')
top_classes, scores = predictor.predict('../test_image/zoe.jpg')
print(f"Top classes: {top_classes}")
print(f"Scores: {scores}")
```
